### 用电量

```python
elec=pd.read_csv("../data/electricity consumption.csv",index_col="province")
elec_cons=elec["2060"]
elec_cons=elec_cons*100000
elec_cons
```

```python
W_pvwind=pd.read_csv("../data/pv and wind power potential for provinces.csv",usecols=["Province","W_Wind","W_PV"])
# W_pvwind=W_pvwind[["Province","W_Wind","W_PV"]]
W_pvwind=W_pvwind.set_index("Province")
W_pvwind[["W_Wind","W_PV"]]=W_pvwind[["W_Wind","W_PV"]]*1000000    #单位：MWh
W_pvwind
```

```python
P_hydro=pd.read_csv("../data/HYtable.csv")
hours=sum([3579,3613,3726,3827,3622,3412])/6
```

```python
P_hydro=P_hydro.set_index("province")
```

```python
W_hydro=P_hydro
W_hydro=W_hydro.rename(columns={'Capacity (MW)':'W_hydro'})
W_hydro["W_hydro"]=P_hydro["Capacity (MW)"]*hours
W_hydro
```

```python
P_nuclear=pd.read_csv("../data/NUtable.csv")
P_nuclear=P_nuclear.rename(columns={"State/Province":"province"})
P_nuclear=P_nuclear.set_index("province")
P_nuclear
```

```python
W_nuclear=P_nuclear
W_nuclear=W_nuclear.rename(columns={'Capacity (MW)':'W_nuclear'})
W_nuclear["W_nuclear"]=P_nuclear["Capacity (MW)"]*365*24*0.9  #0.9的容量因子 
W_nuclear
```

### 各省可再生能源发电量2060

```python
W=W_pvwind.copy()
for i in list(W_hydro.index):
    W.loc[i,"W_hydro"]=W_hydro.loc[i]["W_hydro"]
for j in list(W_nuclear.index):
    W.loc[j,"W_nuclear"]=W_nuclear.loc[j]["W_nuclear"]
W
```

```python
W.loc["National","W_hydro"]=W["W_hydro"].sum()
W.loc["National","W_nuclear"]=W["W_nuclear"].sum()
W
```

```python
W.fillna(0, inplace=True)
```

```python
Energy=W
```

```python
for i in list(elec_cons.index):
    Energy.loc[i,"elec_cons"]=elec_cons.loc[i]
Energy
```

```python
Energy.loc["National","elec_cons"]=Energy["elec_cons"].sum()
Energy
```

```python
Energy["demand"]=Energy["elec_cons"]-Energy["W_hydro"]-Energy["W_nuclear"]
```

```python
Energy["rest"]=Energy["W_Wind"]+Energy["W_PV"]+Energy["W_hydro"]+Energy["W_nuclear"]-Energy["elec_cons"]
```

```python
Energy.to_csv("../data/electricity generation and consumption")  #单位：兆瓦时
```

```python
surplus=Energy["rest"]
surplus.drop(index="National",inplace=True)
surplus  #单位：兆瓦时
```

```python
surplus=surplus/1e8  #单位：亿兆瓦时
surplus
```

```python
surplus=surplus.sort_values()
surplus.to_csv("../data/electricity surplus.csv")
```

```python
#各省最大制氢量
gamma=53
h2=surplus/gamma    #单位：吨
h2
```

```python
h2[h2<0]=0
h2.to_csv("../data/h2_max.csv")
```

```python
alpha=0.5  #实际生产了多少比例的氢
h2_real=h2*alpha
h2_real.to_csv("../data/h2_real.csv")
```

```python
h2_real=pd.read_csv("../data/h2_real.csv",index_col="Province")
```

```python
h2_real
```

```python
h2_total=h2_real.sum().values[0]
h2_total
```

```python
h2_total_indus=float(0.6)*h2_total     #0.6:工业  0.31：交通
h2_total_trans=float(0.31)*h2_total
```
