```python
df1=pd.read_excel("../data/GEM-GGIT-Gas-Pipelines-December-2022.xlsx")
```

```python
df2=df1.loc[(df1["StartCountry"]=="China")&(df1["EndCountry"]=="China")]
```

```python
df2=df2[["PipelineName","SegmentName","Status","StartLocation","EndLocation","Capacity","CapacityUnits","StartState/Province","EndState/Province","WKTFormat"]]
df2
```

```python
df2[(df2["CapacityUnits"]!="bcm/y")&(~df2["CapacityUnits"].isnull())]
#MMSCMD,Million Standard Cubic Metre per Day,*365/1000即转换为bcm
#104cm/d,万立方米/天，*365/100000即转换为bcm
#！！！省内的，不用管
```

#### 补全起始、终点省份缺失值

```python
df_cross=df2[df2["StartState/Province"]!=df2["EndState/Province"]]
df_miss=df_cross[(df_cross["StartState/Province"].isnull())|(df_cross["EndState/Province"].isnull())]
```

```python
drop_index=df_miss.drop(2801,axis=0).index.tolist()
df_cross=df_cross.drop(drop_index,axis=0)
```

```python
df_cross.loc[2801]["StartState/Province"]="Henan"
df_cross.loc[2801]["EndState/Province"]="Shandong"
df_cross.loc[2801]
```

```python
df_cross.loc[2802]["StartState/Province"]="Jiangxi"
df_cross.loc[2802]["EndState/Province"]="Fujian"
df_cross.loc[2802]
```

#### 补全容量，统一容量单位

```python
df_cross["CapacityUnits"]=df_cross["CapacityUnits"].fillna("bcm/y")
```

```python
df_miss_cap=df_cross[(df_cross["Capacity"].isnull())]
df_miss_cap
```

```python
df_cross_new=df_cross[~(df_cross["Capacity"].isnull())]
df_cross_new
```

```python
# Define the function to fit
def linear_func(x, a, b):
    return a * x + b

# Define the data points
x = np.array([2015,2016,2017,2018,2019,2020,2021])
y = np.array([1, 4, 1, 3])

# Fit the data points using the defined function
popt, pcov = curve_fit(linear_func, x, y)

# Print the optimal values for a and b
print(popt)
```
